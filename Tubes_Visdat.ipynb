```python
# Import library
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColumnDataSource, HoverTool, CustomJS, Select
from bokeh.layouts import row, column
from bokeh.io import output_notebook
output_notebook()
```

```python
# Membaca Dataset
data_covid = pd.read_csv('covid-19_Indonesia.csv')
data_covid
```

```python
data_covid.shape
```

```python
# Melihat info data
data_covid.info()
```

```python
#cek missing values
data_covid.isnull().any()
```

```python
# Hapus missing value
data_covid.dropna(inplace=True)
```

```python
data_covid.isnull().sum()
```

```python
#karena ada beberapa atribut yang tidak terlalu penting nilainya maka akan melakukan drop data
data_covid.drop('Longitude', inplace=True, axis=1)
data_covid.drop('Latitude', inplace=True, axis=1)
data_covid.drop('New Cases per Million', inplace=True, axis=1)
data_covid.drop('Total Cases per Million', inplace=True, axis=1)
data_covid.drop('New Deaths per Million', inplace=True, axis=1)
data_covid.drop('Total Deaths per Million', inplace=True, axis=1)
data_covid.drop('Total Deaths per 100rb', inplace=True, axis=1)
```

```python
# cek data 
data_covid
```

```python
# Melihat nama-nama Provinsi
province = data_covid['Location'].unique()
province
```

```python
# Merubah nama kolom
data_covid.rename(columns = {'Total Cases':'Total_Cases'}, inplace = True)
data_covid.rename(columns = {'New Cases':'New_Cases'}, inplace = True)
data_covid.rename(columns = {'Growth Factor of New Cases':'Growth_Factor_of_New_Cases'}, inplace = True)

data_covid.head()
```

```python
# Merubah format tipe data kolom date
data_covid['Date'] = pd.to_datetime(data_covid['Date'])
data_covid.dtypes
```

```python
# Membuat variabel untuk provinsi Banten
df_Banten = pd.DataFrame(data_covid[data_covid["Location"] == 'Banten'])
df_Banten.head()
```

```python
# Membuat variabel untuk provinsi Daerah Istimewa Yogyakarta
df_DIY = pd.DataFrame(data_covid[data_covid["Location"] == 'Daerah Istimewa Yogyakarta'])
df_DIY.head()
```

```python
# Membuat variabel untuk provinsi Jawa Barat
df_Jabar = pd.DataFrame(data_covid[data_covid["Location"] == 'Jawa Barat'])
df_Jabar.head()
```

```python
# Membuat variabel untuk provinsi Jawa Tengah
df_Jateng = pd.DataFrame(data_covid[data_covid["Location"] == 'Jawa Tengah'])
df_Jateng.head()
```

```python
# Membuat variabel untuk provinsi Jawa Timur
df_Jatim = pd.DataFrame(data_covid[data_covid["Location"] == 'Jawa Timur'])
df_Jatim.head()
```

```python
# Menyimpan data pada ColumnDataSource
source_DIY = ColumnDataSource(df_DIY)
source_Banten = ColumnDataSource(df_Banten)
source_Jabar = ColumnDataSource(df_Jabar)
source_Jateng = ColumnDataSource(df_Jateng)
source_Jatim = ColumnDataSource(df_Jatim)
```

```python
# Membuat file Output berbentuk html
output_file('Hasil Analisis Data Covid-19.html', title='Data Covid di Pulau Jawa')
```

```python
# Membuat tooltips dan hover

tooltips = [('Location','@Location'), ('Growth_Factor_of_New_Cases', '@Growth_Factor_of_New_Cases')]
tooltips1 = [('Location','@Location'), ('Growth_Factor_of_New_Cases', '@Growth_Factor_of_New_Cases')]
tooltips2 = [('Location','@Location'), ('Total_Cases','@Total_Cases')]
tooltips3 = [('Location','@Location'), ('New_Cases', '@New_Cases')]
```

```python
# Membuat Figure Analisis
fig = figure(plot_height=600, plot_width=800, x_axis_type='datetime', x_axis_label='Date', y_axis_label='Growth_Factor_of_New_Cases',
             title='Faktor Pertumbuhan Kasus Baru Covid Provinsi di Pulau Jawa')

fig.add_tools(HoverTool(tooltips=tooltips))

fig.line(x='Date', y='Growth_Factor_of_New_Cases', source=source_DIY, legend_label='Daerah Istimewa Yogyakarta', color='red')
fig.line(x='Date', y='Growth_Factor_of_New_Cases', source=source_Banten, legend_label='Banten', color='green')
fig.line(x='Date', y='Growth_Factor_of_New_Cases', source=source_Jabar, legend_label='Jawa Barat', color='Blue')
fig.line(x='Date', y='Growth_Factor_of_New_Cases', source=source_Jateng, legend_label='Jawa Tengah', color='yellow')
fig.line(x='Date', y='Growth_Factor_of_New_Cases', source=source_Jatim, legend_label='Jawa Timur', color='black')

options = [tooltips, tooltips1, tooltips2, tooltips3]
```

```python
# Membuat callback untuk mengupdate tooltips
callback = CustomJS (args=dict(tt=fig.hover, opts=options), code="""
    if (cb_obj.value=='Normal') {
        tt[0].tooltips=opts[0]
    } else if (cb_obj.value=='Growth_Factor_of_New_Cases') {
      tt[0].tooltips=opts[1]
    } else if (cb_obj.value=='Total_Cases') {
      tt[0].tooltips=opts[2]
    } else {
        tt[0].tooltips=opts[3]
    }
    """)
```

```python
# Membuat menu dropdown parameter
stat_select = Select(options=['Total_Cases', 'New_Cases', 'Growth_Factor_of_New_Cases'], title='Macam-Macam Parameter:')
stat_select.js_on_change("value", callback)
```

```python
# Visualisasi
fig.legend.location = "top_left"
fig.legend.click_policy="hide"
show(column(stat_select,fig))
```
